```python
file = codecs.open('List_of_postal_codes_of_Canada:_M', encoding='utf-8')
html = file.read()
soup = BeautifulSoup(html, 'html.parser')

df_canada = (pd.read_html(str(soup.table))[0])

#drop the first row, which is the headings of the table
df_canada.drop(axis=0, index=0, inplace=True)

#then assign the proper column names
column_names = ['Postcode', 'Borough', 'Neighborhood'] 
df_canada.columns = column_names

#get rid of unassigned Boroughs
df_canada = df_canada[df_canada['Borough'] != "Not assigned"].reset_index(drop=True) 
df_canada
```

```python
na_neighborhood= df_canada[df_canada['Neighborhood']=="Not assigned"]
for index in range(len(na_neighborhood)):
    new_indexValue = na_neighborhood.iloc[index]
    new_indexValue['Neighborhood']=new_indexValue['Borough']
    df_canada.iloc[na_neighborhood.index[0]]['Neighborhood'] = new_indexValue['Borough']

df_canada.head()
```

```python
aggregations = {'Neighborhood':lambda x:', '.join(x)}
g = df_canada.groupby('Postcode').agg(aggregations).reset_index()
x = df_canada.drop('Neighborhood',axis=1)
m = pd.merge(x, g, on='Postcode', how='left')
# dropped duplicates
df_torronto = m.drop_duplicates()
df_torronto.to_csv("Toronto_Neighborhood.csv", index=False)
df_torronto 
```
